```python
from selenium import webdriver
import time
import pandas as pd
import os
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service

#url data scietist USA 
url = "https://www.linkedin.com/jobs/search?keywords=Data%20Scientist&location=United%20States&geoId=103644278&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0"

#data engeneer USA
url1 = "https://www.linkedin.com/jobs/data-engineer-jobs?currentJobId=3728707694&position=1&pageNum=0"

#setting up chromedriver using executable
chrome_driver_path = os.path.join(os.getcwd(), "chromedriver")
service = Service(executable_path=chrome_driver_path)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

#open url 
driver.get(url1)
driver.implicitly_wait(10)
```

```python
companyname_list = []
jobs_list = []
final_list = []
useless_list = []
wait = WebDriverWait(driver, 10)


def start_with_me():
    job_html_class = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
    for i in job_html_class:
        single_job = i.text
        useless_list.append(single_job)
    print(len(useless_list))
    return useless_list

def getting_job_list():
    job_html_class = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
    for i in job_html_class:
        single_job = i.text
        jobs_list.append(single_job)
    print(len(jobs_list))
    return jobs_list

def getting_company_name():
    company_html_class = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')
    for j in company_html_class:
        company = j.text
        companyname_list.append(company)
    print(len(companyname_list))
    return companyname_list

def getting_final_list():
    for i in range(625):
        try:
            # Find all job postings
            job_html_class = driver.find_elements(By.CLASS_NAME, 'base-card')

            # Click on the current job posting title
            driver.execute_script("arguments[0].click();", job_html_class[i - 1])

            # Wait for the "Show more" button and click it
            show_more_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Show more, visually expands previously read content above']")))
            driver.execute_script("arguments[0].scrollIntoView(true);", show_more_button)
            driver.execute_script("arguments[0].click();", show_more_button)

            # Wait for the job description to load
            wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'show-more-less-html__markup')))
            job_desc_html_class = driver.find_element(By.CLASS_NAME, 'show-more-less-html__markup')
            
            job_desc = job_desc_html_class.text.strip()
        
            final_list.append(f"Job Posting Description:\n{job_desc}")

            # Go back to the job listings page
            driver.execute_script("window.history.go(-1)")
        except Exception as e:
            final_list.append(f"Job Posting {i} Description: Huston we have a problem")
            print(f"Error getting job description for posting {i}: {str(e)}")

    print(len(final_list))
    return final_list

def scroll_down():
    i = 0
    while i <= 4: 
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        i = i + 1
        try:
            send = driver.find_element(By.XPATH, "//button[@aria-label='See more jobs']")
            driver.execute_script("arguments[0].click();", send)   
            time.sleep(3)      
        except:
            pass
        print(i)
```
